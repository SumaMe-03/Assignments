```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
image = cv2.imread('/content/moon.tif',0)
cv2_imshow(cv2.resize(image,(400,400)))
```

```python
#Contrast Stretching

# Calculate minimum and maximum pixel values
min_val, max_val, _, _ = cv2.minMaxLoc(image)

# Stretch the contrast
contrast_stretched = cv2.normalize(image, None, 0, 255, cv2.NORM_MINMAX)

# Display or save the result
cv2_imshow(contrast_stretched)
```

```python
# Bit Plane slicing
for i in range(8):
    bit_plane = image & (1 << i)
    plt.subplot(4,4,i+1), plt.imshow(bit_plane, cmap='gray')
    plt.title(f'{i}th SLicing')
    plt.axis('off')
plt.show()
```

```python
#gray level slicing

# Define the gray level range to highlight
lower_threshold = 100
upper_threshold = 150

# Apply gray level slicing
gray_level_sliced = cv2.inRange(image, lower_threshold, upper_threshold)
plt.imshow(gray_level_sliced, cmap='gray')
plt.title('Gray level Slicing')
plt.axis('off')
plt.show()
```
