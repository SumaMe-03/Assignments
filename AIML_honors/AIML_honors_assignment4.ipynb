```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("iris.csv")
```

```python
data.head()
```

```python
import seaborn as sns
from sklearn.decomposition import PCA
```

```python
data.describe()
```

```python
sns.heatmap(data.corr())
```

```python
pca = PCA()
```

```python
x_new1 =pca.fit_transform(data.drop(["variety"],axis =1))
x_new1[:5]
```

```python
expl= pca.explained_variance_ratio_
```

```python
expl
```

```python
plt.figure(figsize =(10,6))
plt.bar(range(4),expl, alpha=0.5, align='center', label ="Individual Explained Variance", color ="magenta")
plt.ylabel("Explained Variance Ratio")
plt.xlabel("Principal Component")
plt.legend(loc ="best")
plt.tight_layout()
```

```python
pca  = PCA(n_components=3)
x_new =pca.fit_transform(data.drop(['variety'],axis =1))
x_new[:5]
```
