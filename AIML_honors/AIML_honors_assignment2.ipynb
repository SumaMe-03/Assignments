```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from scipy.stats import skew,kurtosis
import scipy.stats as stats
```

```python
data = pd.read_csv("student_marks.csv")
```

```python
data
```

```python
data = data.iloc[:,:2]
data
```

```python
marks = data['Marks']
marks
```

```python
mean = np.mean(marks)
```

```python
mean
```

```python
sd = np.std(marks, ddof=1)
```

```python
sd
```

```python
median = np.median(marks)
```

```python
median
```

```python
fit = stats.norm.pdf(marks, mean, sd)
```

```python
plt.hist(marks, density = True, color ="blue",ec ="white",)
plt.plot(marks,fit,"go:")
plt.title("Math Scores")
plt.xlabel("Marks")
plt.ylabel("Normal Distribution")

plt.show()
```

```python
skewness = np.mean((marks - np.mean(marks)) ** 3) / np.std(marks) ** 3
kurtosis = np.mean((marks - np.mean(marks)) ** 4) / np.std(marks) ** 4
```

```python
skewness
```

```python
kurtosis
```
